```python
import pandas as pd
import plotly.express as px
from constants import ROOT_DATA_DIR

df = pd.read_parquet(ROOT_DATA_DIR / 'merged_df.parquet')

df['lp_token_supply_future'] = df['lp_token_total_supply'].shift(-30) #3 reward rate +x days after today
df['reward_rate_eth_future'] = df['reward_rate_eth'].shift(-30) 

# df['percent_change_in_supply'] = 100 * (df['lp_token_supply_shifted'] - df['lp_token_total_supply']) / df[['lp_token_supply_shifted', 'lp_token_total_supply']].mean(axis=1)
# df['percent_change_in_reward_rate'] = 100 * (df['reward_rate_eth_shifted'] - df['reward_rate_eth']) / df[['reward_rate_eth', 'reward_rate_eth_shifted']].mean(axis=1)

df['percent_change_in_supply'] = 100 * (df['lp_token_supply_future'] - df['lp_token_total_supply']) / df[['lp_token_total_supply', 'lp_token_supply_future']].mean(axis=1)
df['percent_change_in_reward_rate'] = 100 * (df['reward_rate_eth_future'] - df['reward_rate_eth']) /df[['reward_rate_eth', 'reward_rate_eth_future']].mean(axis=1)
df['elasticity'] = df['percent_change_in_supply'] / df['percent_change_in_reward_rate']

limited_df = df[df['percent_change_in_reward_rate'].abs() > 1]
print(df.shape, limited_df.shape)
px.line(limited_df['elasticity'])
```
