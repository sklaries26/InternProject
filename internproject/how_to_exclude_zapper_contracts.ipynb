```python
import pandas as pd
import plotly.express as px
import warnings 
from internproject.constants import ROOT_DATA_DIR

add_lp = pd.read_parquet(ROOT_DATA_DIR / "stETH_ETH_add_lp.parquet")  # more details on the add liquidity events
remove_lp = pd.read_parquet(
    ROOT_DATA_DIR / "stETH_ETH_remove_lp.parquet"
)  # more details on the remove liquidity events
lp_transfers = pd.read_parquet(ROOT_DATA_DIR / "stETH_ETH_lp_transfers.parquet")

zapper_address = '0x271fbE8aB7f1fB262f81C77Ea5303F03DA9d3d6A'

null_address = "0x0000000000000000000000000000000000000000"
add_liquidity_transfers = lp_transfers[lp_transfers["from"] == null_address].copy()
withdraw_liquidity_transfers = lp_transfers[lp_transfers["to"] == null_address].copy()

add_liquidity_transfers['user'] = add_liquidity_transfers['to']
withdraw_liquidity_transfers['user'] = withdraw_liquidity_transfers['from']
withdraw_liquidity_transfers['value'] = withdraw_liquidity_transfers['value']* -1
transfer_df = pd.concat([add_liquidity_transfers, withdraw_liquidity_transfers])


# you can't check for cumsol_col for negatives because the qunaity of lp tokens in the stETH:ETH pool increases 
# because of the stETH rebasing
# This a a quirk of the curve stETH:ETH because of the stETH rebasing
# instead you look for users that have withdrawn liquidity BEFORE they have entered. 
# or have withdrawn but never have entered
# Cumulative sum of lp transfers will work for identifying whales but know that it could be ~5% off. 
```

```python
users_first_add = transfer_df[transfer_df['value'] > 0].groupby('user')[['timestamp']].min()
users_first_add.columns = ['date_added']
users_first_remove = transfer_df[transfer_df['value'] < 0].groupby('user')[['timestamp']].min()
users_first_remove.columns = ['date_removed']
users_df = users_first_add.join(users_first_remove, how='outer')
users_df['date_removed'] = users_df['date_removed'].fillna(pd.to_datetime('2050-01-01')) # far in the future
users_df
```

```python
users_who_never_added = users_df[users_df['date_added'].isna()].index
users_who_removed_before_they_added = users_df[users_df['date_added'] > users_df['date_removed']].index
len(users_who_never_added), len(users_who_removed_before_they_added)
```

```python
transfer_df[transfer_df['user'] == users_who_never_added[0]][['from', 'to', 'value']]
# They never added but did withdraw, best guess is the used zapper
```

```python
transfer_df[transfer_df['user'] == users_who_removed_before_they_added[0]][['from', 'to', 'value']]
# this person removed before they added, best guess is they used the zapper
```

For you analysis you can exclude `users_first_add` and `users_who_never_added`. That will miss some people who used a the zapper and also the normal method but it should be good enough for now.
